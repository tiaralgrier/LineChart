2. Load in the dataset def_area_2004_2019.csv using pandas. Check out the first few lines to make sure you know how the data is structured.

```python
data = pd.read_csv('def_area_2004_2019.csv')
data.rename(columns={
    'Ano/Estados': 'Year_of_occurrence',
    'AC': 'Acre',
    'AM': 'Amazonas',
    'AP': 'Amapa',
    'MA': 'Maranhao',
    'MT': 'Mato_Grosso',
    'PA': 'Para',
    'RO': 'Rondonia',
    'RR': 'Roraima',
    'TO': 'Tocantins'
    }, inplace=True) # Rename columns in English
data.head()
```

3. Write the code to make a line chart with Ano (Year) on the x-axis and loctaion AC on the y-axis, then print the plot.

```python
plt.plot(data.Year_of_occurrence, data.Acre)

plt.show()
```

4. Write the code to add lines to the graph for the other eight locations: Amazonas, Amapa, Maranhao, Mato Grosso, Para, Rondonia, Roraima, and Tocantins.

```python
plt.plot(data.Year_of_occurrence, data.Amazonas)
plt.plot(data.Year_of_occurrence, data.Amapa)
plt.plot(data.Year_of_occurrence, data.Maranhao)
plt.plot(data.Year_of_occurrence, data.Mato_Grosso)
plt.plot(data.Year_of_occurrence, data.Para)
plt.plot(data.Year_of_occurrence, data.Rondonia)
plt.plot(data.Year_of_occurrence, data.Roraima)
plt.plot(data.Year_of_occurrence, data.Tocantins)

plt.show()
```
